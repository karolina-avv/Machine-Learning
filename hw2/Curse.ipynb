```python
import numpy as np
import matplotlib.pyplot as plt

def sample_gaussian(m, n_samples=100):
    radii = []
    for _ in range(n_samples):
        sample = np.random.normal(size=m)
        radius = np.linalg.norm(sample)
        radii.append(radius)
    return np.mean(radii), np.std(radii)

m_values = range(1, 41)
mean_radii = []
std_radii = []

for m in m_values:
    mean_radius, std_radius = sample_gaussian(m)
    mean_radii.append(mean_radius)
    std_radii.append(std_radius)

plt.figure(figsize=(10, 6))
plt.plot(m_values, mean_radii, label='Mean Radius')
plt.plot(m_values, std_radii, label='Standard Deviation of Radius')
plt.xlabel('Dimension (m)')
plt.ylabel('Value')
plt.title('Mean and Standard Deviation of Radii in Gaussian Samples')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def gaussian_pdf(x, sigma, m):
    return 1 / np.sqrt((2 * np.pi * sigma**2) ** m) * np.exp(-np.linalg.norm(x)**2 / (2 * sigma**2))

def point_on_sphere(m, r):
    # Generate a random point on the unit sphere in m-dimensional space
    x = np.random.normal(size=m)
    x /= np.linalg.norm(x)
    # Scale the point by r to lie on the sphere with radius r
    return r * x

# Parameters
m = 10  # Dimensionality
sigma = 1.0  # Standard deviation
r = 2.0  # Radius of the sphere

# Calculate PDF at the origin
pdf_origin = 1 / np.sqrt((2 * np.pi * sigma**2) ** m)

# Generate a random point on the sphere
point = point_on_sphere(m, r)

# Calculate PDF at the random point
pdf_point = gaussian_pdf(point, sigma, m)

print("PDF at the origin:", pdf_origin)
print("PDF at a random point on the sphere:", pdf_point)
```
